```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data_26_4\Когортный анализ.xlsx")
df.head()
```

```python
# Проверка на наличие пропусков
print("1)", df[df['user_id'].isna() == True])
print("2)", df[df['date'].isna() == True])
print("3)", df[df['price'].isna() == True])
print("4) price=0:", df[df['price'] == 0] )
```

```python
def change_day(date):
    return dt.datetime(date.year,date.month,1)

def get_date(df,column):
    return df[column].dt.year, df[column].dt.month, df[column].dt.day

df["month_year"] = df["date"].apply(change_day)
```

```python
df['cohort_month'] = df.groupby('user_id')['month_year'].transform('min')
```

```python
date_year,date_month,_ = get_date(df,'date')
cohort_year,cohort_month,_ = get_date(df,'cohort_month')
```

```python
year_diff = date_year-cohort_year
month_diff = date_month-cohort_month
```

```python
df['cohort_index'] = year_diff * 12 + month_diff + 1
df.head()
```

## Когортный анализ для количества покупателей

```python
data1 = df.pivot_table(index='cohort_month',
                      columns='cohort_index',
                      values='user_id',
                      aggfunc='nunique')
data1
```

```python
first_column = data1.iloc[:,0]
```

```python
retention = data1.div(first_column,axis=0)
retention.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('Когортный анализ для количества покупателей')
sns.heatmap(data=retention,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlOrRd")
plt.show()
```

**Вывод:** количество оставшихся во втором периоде клиентов не привышает 40%, на третий месяц оно составляет 13-22%. Лучшие результаты показывает 1 когорта пользователей, в следующих процент меньше по сравнению с ней. Показатели можно повысить введением акций или улучшением обслуживания клиентов.

От других групп отличается отток клиентов, привлечённых в феврале и апреле 2020 года. Возможно, что в эти периоды была привлечена другая категория пользователей или внесены изменения в продукт.

## Когортный анализ для сумм покупок

```python
data2 = df.pivot_table(index='cohort_month',
                      columns='cohort_index',
                      values='price',
                      aggfunc=sum)
data2
```

```python
first_column = data2.iloc[:,0]
```

```python
retention = data2.div(first_column,axis=0)
retention.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('Когортный анализ для сумм покупок')
sns.heatmap(data=retention,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

**Вывод:** среди других выбивается когорта февраля 2020 года. Обратив внимание на процент оставшихся клиентов в группе с 6 периода, могу заметить что показатель стал уменьшаться более интенсивно, чем в других группах.
